```python
#gender cleaning from 4 to 2 (female, male)
def clean_gender(x):
    x_lower = x.lower()
    if x_lower == 'fem':
        xf = x_lower.replace('fem','female')
        return xf
    else:
        return x_lower
proyect_data['gender'] = proyect_data['gender'].apply(clean_gender)

#first word to capital letter
def to_capitalize(x):
    return x.capitalize()
proyect_data['gender'] = proyect_data['gender'].apply(to_capitalize)
proyect_data

#comprobation
#proyect_data['gender'].value_counts()
```
